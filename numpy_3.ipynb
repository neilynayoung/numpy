```python
xarr = np.array([1.1,1.2,1.3,1.4,1.5])
yarr = np.array([2.1,2.2,2.3,2.4,2.5])
cond = np.array([True,False,True,True,False])  #조건
```

```python
#조건에 따라서 데이터 담기(python 방식)
result = [(x if c else y) for x,y,c in zip(xarr,yarr,cond)]   
#3항연산자, cond에 따라서 true면 x를 출력하고 false면 y출력
#zip은 하나로 묶는 명령어
#zip->for->if
result
```

```python
#numpy의 where 기능(조건, 데이터1, 데이터2) / 조건에 따라서 true면 데이터1, false면 데이터1
result = np.where (cond, xarr, yarr)
result
```
