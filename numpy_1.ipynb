```python
#두개가 같은 내용 다른방식

my_arr = np.arange(1000000)  #numpy방식
%time for _ in range(10): my_arr2 = my_arr*2  
# '%time'은 그 한줄 코드를 실행하는데 걸린 시간을 알 수 있는 함수, my_arr에 2를 곱한것
#  _  은 해당 값이 필요하지 않거나 사용되지 않는 값을 _에 할당하기만 하면된다.
```

```python
my_list = list(range(1000000))
%time for _ in range(10): my_list2 = [x*2 for x in my_list] #my_list에 2를 곱한 것
```
